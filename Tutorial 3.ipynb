# 1. Shannon's information entropy

## 1.1. Interpretation

Entropy originates from statistical mechanics and intends to measure how "random" a physical system is. Shannon borrows the idea and defines information entropy, which becomes the cornerstone for his theory of communication. The **Shannon entropy** of a distribution $P(x)$ is the expected amount of information in an event drawn from that distribution, defined as:

$$
H(X)=\left\langle{\log\left(\frac{1}{P(x)}\right)}\right\rangle 
$$

Why does $\frac{1}{P(x)}$, the inverse of the probability of the event $x$, represent the information of the event? One may interpret this quantity as the event's "information value":

- $P(x)$ is large - The event happens again and again, and you will lose interest in observing it. So this event is not valuable.
- $P(x)$ is small - If a rare event suddenly happens, it surely catches your attention. You may want to get as many data as possible from this observation, so this event is very valuable.

(i.e. Rare is expensive. 物以罕為貴.) 

In a system of events with larger $H(X)$, it contains more events with higher rarity. We can use such system to carry more kinds of information. For example, we can only assign total two meanings to a coin (e.g. head = go to school, tail = not go to school), but total 6 meanings to the faces of a dice. Thus a dice is a better information carrier than a coin. Therefore, you may interpret that **_$H(X)$ represents the variance of information that a system can carry_**. In some text it is simply described as the "uncertainty" of a system, fitting the name entropy. 




## 1.2. Formulation

Depends on the type of variable of the system, 

- For a discrete random variable $X$, which picks any values from $\{x_1,x_2,\dots, x_n\}$ with probabilities $\{p_1,p_2,\dots, p_n\}$. It becomes

    $$
    H(X)\overset{\text{disc. }X}\equiv -\sum_i p_i \log(p_i) \ ,
    $$

- For a continuous variable $X$ distributed with $f_X(x)$, it is defined as

    $$
    H(X) \overset{\text{cont. }X} \equiv - \int f_X(x) \log f_X(x) \mathrm{d}x \ .
    $$

    It is sometimes referred to as the **differential entropy**. Shannon comes up with this formula simply by replacing summation with integration. This definition usually works, but it creates anomaly when one tries to transform variables, i.e. replacing $X$ with $Y=g(X)$ for some function $g$. This failure arises because it is dimensionally wrong: Unit of $f_X(x)\sim \frac{1}{\mathrm{d}x}$, but logarithm requires a dimensionless argument. 


(In this tutorial, we will focus on the definition with discrete variables only.)




## 1.3. Unit 
Note that one can use any base because a change in base only rescales the entropy. These bases are commonly used:

- Base = $2$ - The unit is a **bit**. 
- Base = $e$ - Easier for calculus. The unit is a **nat**. 
- Base = $10$ - Rather rare to see. The unit is called a **dit**.

#### Example: Tossing a coin 1

What is the entropy of the event "tossing head 10 times in a series"? Assume the probability to be a head is $p$

**Solution.** The event may or may not be rare, depends on the coin:

- For a fair coin, $p=0.5$. Probability of this event is $\frac{1}{2^{10}}$, so the entropy of this event is $\log_2 2^{10} = 10$ (relatively valuable)
- For a 2-head coin, $p=1$. Probability of this event is $1$, so the entropy is $\log_2 1 = 0$ (no value at all).


#### Example: Tossing a coin 2

What is the expected entropy of tossing a coin whose probability to be a head is $p$?

**Solution**. The coin gives head by probability $p$ and tail by probability $1-p$. The expected entropy is thus

$$
H_\text{bin}(X)=-p\log_2 p-(1-p)\log_2 (1-p) \ .
$$

Since a coin is a binary system, we can intuitively choose base = $2$. This is in fact so called the **binary entropy function**, the formula of the entropy of all binary variables.

# 2. Theory of communication

## 2.1. Terminology

A communication consists of these components:

- **Transmitter** $X$ = The agent who convert a meaningful message (information) to a signal.
- **Receiver** $Y$ = The agent who reconstruct the signal back to a meaningful message (information).
- **Information** = The abstract concept we want to pass from $X$ to $Y$.
- **Signal** = The series of symbols that is physically sent from $X$ and received by $Y$.
- **Encoding** = How the abstract information is translated into symbols $\{x_1, x_2,\cdots x_m\}$ at the transmitter side.
- **Decoding** = How the symbols $\{y_1, y_2,\cdots y_n\}$ are reconstructed back to the abstract information at the receiver side.
- **Channel** = The medium for transmitting the signal from transmitter to receiver.

We consider the simplest model - **discrete memoryless channel model**. 
- By discrete: Signal sent by $X$ are composed of discrete values.
- By memoryless: Interpretation by $Y$ only depends on the symbol received at the moment, not related to symbols received in the past. 

In the ideal case, we want every bit sent by $X$ to be correctly received by $Y$. However, physical errors can occur and $Y$ may sometimes mis-interpret the message. For example, electronics signal $x_i=\{\text{high},\text{low}\}$ can become $y_j=\{\text{high},\text{low},\text{undertermined}\}$ when arriving $Y$, if noise is added.

One can define the **channel transition probabilities** as the conditional probabilities $p_{ij} = P(y_j|x_i)$, i.e. the probability of interpreting into $y_j$ when the symbol $x_i$ is sent. In the ease of computation, it is common to pack all the $p_{ij}$ into one matrix, the **channel matrix**, $P(Y|X)$. E.g. 

$$ 
P(Y|X)= 
\begin{bmatrix}
P(y_1|x_1) & P(y_1|x_2) \\
P(y_2|x_1) & P(y_2|x_2) \\
P(y_3|x_1) & P(y_3|x_2)
\end{bmatrix}
=
\begin{bmatrix}
p_{11} & p_{12} \\
p_{21} & p_{22} \\
p_{31} & p_{32}
\end{bmatrix}
$$

So that the conditional probability rule can be made into a compact form.

$$
P(Y) = \begin{bmatrix} P(y_1) \\ P(y_2) \\ P(y_3) \end{bmatrix}
=
\begin{bmatrix}
P(y_1|x_1) & P(y_1|x_2) \\
P(y_2|x_1) & P(y_2|x_2) \\
P(y_3|x_1) & P(y_3|x_2)
\end{bmatrix}
\begin{bmatrix} P(x_1) \\ P(x_2) \end{bmatrix}
=
P(Y|X)P(X) 
$$

If $P(X)$ is written as a diagonal matrix, each entry in the result matrix has the form $P(y_j|x_i)P(x_i) = P(x_i, y_j)$, which is the joint probability of transmitting $x_i$ and receiving $y_j$. The matrix is called the **joint probability matrix**, $P(X,Y)$.

$$
P(X,Y) = 
\begin{bmatrix}
P(y_1|x_1)P(x_1) & P(y_1|x_2)P(x_2) \\
P(y_2|x_1)P(x_1) & P(y_2|x_2)P(x_2) \\
P(y_3|x_1)P(x_1) & P(y_3|x_2)P(x_2)
\end{bmatrix}
=
\begin{bmatrix}
P(y_1|x_1) & P(y_1|x_2) \\
P(y_2|x_1) & P(y_2|x_2) \\
P(y_3|x_1) & P(y_3|x_2)
\end{bmatrix}
\begin{bmatrix} P(x_1) & 0\\ 0& P(x_2) \end{bmatrix}
=
P(Y|X)\cdot\text{diag}(P(X))
$$

## 2.2 From probability to entropy

From the above definitions to probabilities, we can define several entropy functions. Here $\odot$ denotes the elementwise product between two matrix. 

- **Entropy of the source and recieved signals**, i.e. the average variance/uncertainty of information they can carry.

 $$
\begin{align*}
H(X) &= -\sum_{i=1}^m P(x_i)\log P(x_i) 
= \begin{bmatrix} P(x_1) & \cdots & P(x_m) \end{bmatrix} 
\begin{bmatrix} \log P(x_1) \\ \vdots \\ \log P(x_m) \end{bmatrix} 
= -\sum \Big[P(X) \odot\log (P(X))\Big]
\\[0.5em]
H(Y) &= -\sum_{j=1}^n P(y_j)\log P(y_j)
= \begin{bmatrix} P(y_1) & \cdots & P(y_n) \end{bmatrix} 
\begin{bmatrix} \log P(y_1) \\ \vdots \\ \log P(y_n) \end{bmatrix} 
= -\sum \Big[P(Y)\odot \log(P(Y))\Big]
\end{align*}
$$

- **Conditional entropy** of $Y$ upon $X$, and vice versa. For example, $H(Y|X)$ is the variance/uncertainty of information appears at $Y$ even $X$ is known. They basically describe how noisy the channel since higher variance = more possibility = harder to choose the correct one:

    - If $H(Y|X)$ is high, then person at $Y$ cannot tell what symbol $X$ really sent to him since the received message is mixed with noise. 
    - If $H(X|Y)$ is high, then person at $X$ will suffer more in deciding which symbol to send to $Y$ such that $Y$ will receive the correct symbol, because noise will distort his input.
 
 <br>Note that you need the Baysian inference formula to compute $P(X|Y)$ from $P(Y|X)$.
 
 $$
\begin{align*}
H(Y|X) &= -\sum_{i=1}^m \sum_{j=1}^n P(x_i,y_j) \log P(y_j|x_i) = -\sum \Big[P(X,Y) \odot \log(P(Y|X))\Big]
\\
H(X|Y) &= -\sum_{i=1}^m \sum_{j=1}^n P(x_i,y_j) \log P(x_i|y_j) = -\sum \Big[P(X,Y)^\intercal \odot \log(P(X|Y))\Big]
\end{align*}
$$

 

- **Joint entropy** is the average variance/uncertainty of the whole communication system. 

 $$
\begin{align*}
H(X,Y) &= -\sum_{i=1}^m \sum_{j=1}^n P(x_i,y_j) \log P(x_i, y_j) = - \sum \Big[P(X,Y) \odot \log(P(X,Y))\Big] \\
&= H(X|Y) + H(Y) \\
&= H(Y|X) + H(X)
\end{align*}
$$

## 2.3. Mutual information

To measure the performace of the channel, we define the **mutal information** as 

$$
I(X;Y) = H(X) - H(X|Y)
$$

Recall the definitions from above sections, 

- $H(x)$ = Average variance/uncertainty of the message that can be carried by the system at $X$. Higher = Possible to have richer content.
- $H(X|Y)$ = Average variance/uncertainty from noise's distortion that prevent $X$ sending the correct symbol to $Y$. Lower = Cleaner channel.

Therefore $I(X;Y)$ can be interpret as the average amout of correct information that can be sent from $X$ to $Y$. The maximum value of $I(X;Y)$ among all possible $P(X)$ is called the **channel capacity**.

$$
C=\max_{P(X)} I(X;Y)
$$

Mathematically, $I(X;Y)$ is symmetric about $X$ and $Y$. These expressions are equivalent to each other:

$$ 
\begin{align*}
I(X;Y) &= H(X) - H(X|Y) \\
&= H(Y) - H(Y|X) \\
&= H(X,Y) - H(X|Y) - H(Y|X) \\
&= H(X) + H(Y) - H(X,Y) \\
&= \sum_{i=1}^m \sum_{j=1}^n P(x_i, y_j) \log\left(\frac{P(x_i,y_j)}{P(x_i)P(y_j)}\right)
\end{align*}
$$

#### Example: Clean channel vs noisy channel

What is the mutual information for a channel if 

1. Symbols received at $Y$ is totally independent from the symbols sent by $X$?
2. Symbols received at $Y$ is always the same as the symbols sent by $X$?

**Solution.** 

1. $Y$ is independent from $X$, so $P(x,y) = P(x)P(y)$. From the formula, it gives $I(X;Y) = 0$. i.e. No information can be sent.

2. $Y$ always follow $X$, so the conditional probability $P(x|y) = 1$ for $x=y$ and $0$ otherwise. So $P(x,y) = P(x) = P(y)$. From the formula

 $$
\begin{align*}
I(X,Y) &= \sum_{i=1}^m \sum_{j=1}^n P(x_i, y_j) \log\left(\frac{P(x_i,y_j)}{P(x_i)P(y_j)}\right) \\
&= \sum_{i=1}^n  P(x_i) \log\left(\frac{1}{P(x_i)}\right) \\
& = H(X) = H(Y) = H(X,Y)
\end{align*}
$$

 As expected, the channel size equals to the variance of the encoding scheme by $X$.



## 2.4. Communication channel in practice

### 2.4.1. Binary symmetric channel

The simplest model of communication developed by Shannon is the binary symmetric channel (BSC). The features are:

- Sender side: Send out $X=0$ with a probability $\alpha$ and $X=1$ otherwise. 
- Receiver side: Correctly receives $Y=X$ with a probability $p$ but accidentally $Y=1-X$ otherwise. 

<figure style="text-align: center">


    
<figcaption> <b>Fig. 3</b> Binary symmetric channel (BSC).  </figcaption>
</figure>

The channel is symmetric because the probability of a correct reception does not depend on the input. To be clear, we can write down

- $P(X=0) = \alpha$,
- $P(X=1) = 1-\alpha$,
- $P(Y=0|X=0) = P(Y=1|X=1) = p$,
- $P(Y=0|X=1) = P(Y=1|X=0) = 1-p$

What is the mutual information $I(X;Y)$? There are only a few terms, so let's compute directly.

- For $H(Y)$, first compute the two $P(Y)$:

 $$
\begin{align*}
P(Y=0) &= P(Y=0|X=1)P(X=1)+P(Y=0|X=0)P(X=0) \\
&= (1-p)(1-\alpha) + p\alpha \\
P(Y=1) &= P(Y=1|X=1)P(X=1)+P(Y=1|X=0)P(X=0) \\
&= p(1-\alpha) + (1-p)\alpha
\end{align*}
$$
 
 Then 
 $$
 \begin{align*}
 H(Y) &= -\sum_{j=1}^2 P(y_j)\log(P(y_j)) \\
 &= -[(1-p)(1-\alpha) + p\alpha]\log[(1-p)(1-\alpha) + p\alpha] - [p(1-\alpha) + (1-p)\alpha]\log[p(1-\alpha) + (1-p)\alpha]
 \end{align*}
 $$
 
- For $H(Y|X)$, we need $P(Y|X)$ and $P(X,Y)$. To be clear, list out their matrix forms first

 $$
 \begin{align*}
P(Y|X) &= \begin{bmatrix} p & 1-p \\ 1-p & p \end{bmatrix} \\
P(X,Y) &= P(Y|X)P(X) = \begin{bmatrix} p & 1-p \\ 1-p & p \end{bmatrix}\begin{bmatrix} \alpha & 0 \\ 0 & 1-\alpha \end{bmatrix} 
= \begin{bmatrix} p\alpha & (1-p)(1-\alpha) \\ (1-p)\alpha & p(1-\alpha) \end{bmatrix}\\
\end{align*}
$$

 Then $H(Y|X)$ is the elementwise sum-product between $P(X,Y)$ and $\log(P(Y|X))$.
 
 $$
\begin{align*}
H(Y|X) &= -p\alpha \log(p) -(1-p)(1-\alpha)\log(1-p) - (1-p)\alpha\log(1-p) - p(1-\alpha)\log(p) \\
&= -p\log(p) - (1-p)\log(1-p)
\end{align*}
$$

Then $I(X;Y) = H(Y) - H(Y|X)$. To get the channel's capacity, we need to first solve $\frac{\mathrm{d}}{\mathrm{d}\alpha} I(X;Y)=0$ for the optimal $\alpha^*$ and then substitute back.

$$
\begin{align*}
\frac{\mathrm{d}}{\mathrm{d}\alpha}I(X;Y) &= \frac{\mathrm{d}}{\mathrm{d}\alpha}H(Y) \\
&= (2p-1)\log\left(\frac{p(1-\alpha)+(1-p)\alpha}{p\alpha+(1-p)(1-\alpha)}\right) = 0\\
\alpha &= 0.5
\end{align*}
$$

So the channel capacity is $$C = \max_{\alpha=0.5}I(X;Y) = \log 2 - H(Y|X) = \log 2 + p\log p+ (1-p)\log(1-p)$$

# 3. Data compression 

## 3.1. Shannon's theorem

### 3.1.1. Source-information rate

We can start with the following definitions:

- **Source information rate** $R_s$:
    - How fast an information can be send out by sender $X$. i.e. How much information can $X$ send out per unit time.
    - The thing we would like to maximize.
    - Measured by (bit/sec).


- **Symbol rate** $r$
    - How fast can a symbol be sent out by $X$. 
    - Determined by physical limit of hardware, e.g. by bandwidth of a network.
    - Measured by (symbol/sec).


- **Entropy** $H(X)$ 
    - How much abstract information is carried by one symbol on average, i.e. dependent on the **encoding scheme**. 
    - Determined by entropy of the information carrier, i.e. higher variance/uncertainty = can carry more meanings per symbol.
    - Measured by (bit/symbol)


Their relation is simply the equation 

$$R_s = rH(X)$$ 

To maximize $R_s$, we can work on maximizing both $r$ and $H(X)$. 
- $r$ can be improved if we are electronics engineers. 
- $H(X)$ can be improved if we are mathematicians / computer scientists. 

### 3.1.2. Shannon's theorem

$R_s$ only concerns how fast an information can be sent out of the source, but not related to if the information can be received correctly by the receiver. Although we can send out information by entropy $H(X)$, only $C = \max I(X;Y) = \max (H(X)-H(X|Y))$ at max can be guarenteed to be received correctly, while the remaining is possibly polluted by noise. 

In other words, $C$ is the upper bound of information that can be carried by one symbol, and so $rC$ is the maximum rate of information that can be sent out correctly through the channel. As long as the rate of information required to be send out is less than $rC$, there must be some encoding scheme $H^*(X)<C$ such that the information can be sent with neligible error - This is what Shannon's theorem means.

**Note:** There are in fact quite a few theorems in the name of Shannon. Here we are refering to Shannon's [noisy-channel coding theorem](https://en.wikipedia.org/wiki/Noisy-channel_coding_theorem).

### 3.2.1. Word extension

Consider an abstract piece of information originally being encoded by symbols $\{A,B,C...\}$. We can further compress the number of symbols being used by extending the set of symbols into words, i.e. permutation of symbols. For example, if the message is originally encoded with two symbols $\{A,B\}$ only, 

$$\text{msg} = ABAABBAABAABBBAABBB$$

The word extension are as follow: 

- 1st order extension = $\{A, B\}$. This is just the original message.

- 2nd order extension = $\{AA, AB, BA, BB\} \rightarrow \{0,10,110,111\}$

    $$\text{msg} = AB|AA|BB|AA|BA|AB|BA|AB|BB \rightarrow 10|0|111|0|10|110|10|111$$
    
- 3rd order extension = $\{AAA, AAB, ABA, ABB, BAA, BAB, BBA, BBB\} \rightarrow \{P,Q,R,S,T,U,V,W\}$

    $$\text{msg} = ABA|ABB|AAB|AAB|BAA|BBB \rightarrow RSQQTW$$

It is intuitive that if we can expand the number of symbols in the set (like the 3rd order extension above, using 8 symbols), the length of the message can be greatly reduced. However this is not feasible especially in computer, since computation is restricted to two symbols 0 and 1. So in practice, data compression would be like the 2nd order extension above, by converting the original words of into new words of variable length. (Count it yourself: How many $\{A,B\}$ and $\{0,1\}$ are there?)

### 3.2.2. Code efficiency

Suppose we have an ensemble of information (information in certain distribution, not just a specific piece) and it is encoded through a set of symbols/words. The measure to the goodness of the encoding scheme is by its code efficiency:

$$
\text{efficiency} = \frac{H(X)}{\bar{L}}
$$

- $H(X) = -\sum_i p(x_i)\log_2 p(x_i)$ = entropy, i.e. the average amount of information can be carried per word. Here $p(x_i)$ = probability of occurence of the word $x_i$. The log in entropy is usually base 2, i.e. using bit as unit, but change it if you want to use other unit. 

- $\bar{L} = \sum_i p(x_i)l(x_i)$ = The average number of symbol per word. Here $l(x_i)$ = length of the word $x_i$. 

So it is simply a measure to the average amount of information carried per symbol. Its value is between 0 and 1, guarenteed by the [Shannon's source coding theorem](https://en.wikipedia.org/wiki/Shannon%27s_source_coding_theorem).

Informally speaking, you may think of the situation: if we have a piece of information originally being encoded such that the probabilities of occurence of the words differ very much, this encoding scheme is having an "excess capacity" in carrying the information (i.e. 大才小用) and is described as "low efficiency". On the other hand, it is theoretically possible to find an encoding scheme for the same information but need fewer symbols on average.  

#### Example: n$^\text{th}$ order extension

Suppose we are given a piece of information, which when encoded using two symbols of length 1, i.e. $\{A,B\}$, will result in $p\%$ of symbols being $A$ and $1-p\%$ of the symbols being $B$. Compare the efficiency with the following encoding schemes to the first 2 extension:

- $\{AA, AB, BA, BB\} \rightarrow \{0,10,110,111\}$

- $\{AAA, AAB, ABA, BAA, ABB, BAB, BBA, BBB\} \rightarrow \{0,100,101,110,11100,11101,11110,11111\}$


**Solution.** The terms for the original scheme (1st order) are

- $H(X) = -p\log_2 p - (1-p)\log_2 (1-p)$
- $\bar{L}=1$.
    
For the $n^\text{th}$ order extension, the amount of information it can carry is proportional to $n$ compare to the original symbol set (state without proof), i.e.

$$H(X) \rightarrow H(X^n) = nH(X) = -n[p\log_2 p - (1-p)\log_2 (1-p)]$$ 

While the average number of symbol per word can be found by tabling:

- 2nd order - $\bar{L} = p^2\times 1 + p(1-p)\times 2+p(1-p)\times 3+(1-p)^2\times 3 = -p^2-p+3$

|Permutation|Probability|New word|Word length|
|:---:|:---:|:---:|:---:|
|AA|$p^2$|0|1|
|AB|$p(1-p)$|10|2|
|BA|$p(1-p)$|110|3|
|BB|$(1-p)^2$|111|3|

- 3rd order - $\bar{L} = p^3\times 1 + p^2(1-p)\times 3\times 3+p(1-p)^2\times 5\times 3+(1-p)^3\times 5 = 2p^3-6p^2+5$

|Permutation|Probability|New word|Word length|
|:---:|:---:|:---:|:---:|
|AAA|$p^3$|0|1|
|AAB|$p^2(1-p)$|100|3|
|ABA|$p^2(1-p)$|101|3|
|BAA|$p^2(1-p)$|110|3|
|ABB|$p(1-p)^2$|11100|5|
|BAB|$p(1-p)^2$|11101|5|
|BBA|$p(1-p)^2$|11110|5|
|BBB|$(1-p)^3$|11111|5|

Plotting the efficiency against $p$ will result in the following graph. We can see that when $p$ is high, the efficency is higher with the higher order encoding, i.e. they are better schemes in terms of "less wasting" of the "carrying capacity" per symbols.

## 3.3. Instantaneous code (Optional reading)

(There was a corresponding lecture in Prof. Szeto's old course but no longer in our syllabus.)

Instantaneous code are coding scheme which one can decode each codeword in sequence, without using spaces or reference to succeeding code symbols. Here introduces two basic instantaneous encoding schemes. 

- **Shannon-Fano Code** - High efficiency, but not optimal.
- **Huffman Code** - Optimal, i.e. minimum number of symbol per word.

### 3.3.1. Shannon-Fano Code

Shannon-Fano code consists of 4 steps: 

1. Sort the original words in descending order of their probability.
2. Draw a line to split the list into 2, such that the two parts are (approximately) equally probable.
3. Assign/extend the first partition with code word $0$ and second partition with code word $1$.
4. Repeat step 2-3 to each partition until spliting is not possible.

Note that the encoding scheme may not be unique. 

#### Example: Shannon-Fano Code

Suppose the message is originallly encode through a set of 5 symbols $\{P,Q,R,S,T\}$ with probaiblity $\{0.4,0.2,0.2,0.15,0.05\}$. Construct the Shannon-Fano encoding scheme. 

**Solution.** 

1. The first split can occur either between $\{P, Q\}$ (into 4:6) or $\{Q,R\}$ (into 6:4). Here the latter case is chosen.

|Word|Prob.|1st split|
|:---:|:---:|:---:|
|P|0.4|0|
|Q|0.2|0|
|---|---|---|
|R|0.2|1|
|S|0.15|1|
|T|0.05|1|


2. Only 1 split can occur in the upper partition; while the split to the lower partition should occur between $\{R,S\}$

|Word|Prob.|1st split|2nd split|
|:---:|:---:|:---:|:---:|
|P|0.4|0|0|
|---|---|---|---|
|Q|0.2|0|1|
|---|---|---||
|R|0.2|1|0|
|---|---|---|---|
|S|0.15|1|1|
|T|0.05|1|1|


3. Finally the perform the last split between $\{S,T\}$.

|Word|Prob.|1st split|2nd split|3rd split|
|:---:|:---:|:---:|:---:|:---:|
|P|0.4|0|0||
|---|---|---|---||
|Q|0.2|0|1||
|---|---|---|||
|R|0.2|1|0||
|---|---|---|---||
|S|0.15|1|1|0|
|---|---|---|---|---|
|T|0.05|1|1|1|

So one of the Shannon-Fano coding scheme is $\{P,Q,R,S,T\} \rightarrow \{00,01,10,110,111\}$. You can work out the other option by choosing the first case in step 1. 

### 3.3.2. Huffman code

Huffman code consists of 2 main steps:
- Reduction (something like forward propagation):
    1. Sort the original words in descending order of their probability.
    2. Combine the two least probable words into 1 group, rename the group and set the probability of the group as the sum of the original two words.
    3. Sort the words with the new renamed group again. Repeat step 2-3 until two groups are left.
    
- Splitting (something like backword propagation):
    1. Assign code word 0 and 1 to the two remaining groups.
    2. Trace back each splitting and extend the code word with 0 or 1 to the two groups.
    3. Repeat step 1-2 until all original words are recovered. 
    
Obviously the encoding scheme is not unique at all.


#### Example: Huffman code

Consider the previous example again, which the message is originallly encode through a set of 5 symbols $\{P,Q,R,S,T\}$ with probaiblity  $\{0.4,0.2,0.2,0.15,0.05\}$ . Construct a Huffman encoding scheme.

**Solution.**

1. Reduction = joint the two least probable words

    \begin{align*}
    \{\underbrace{P}_{0.4},\underbrace{Q}_{0.2},\underbrace{R}_{0.2},\underbrace{S}_{0.15},\underbrace{T}_{0.05}\} 
    &\quad\rightarrow\quad
    \{\underbrace{P}_{0.4},\underbrace{Q}_{0.2},\underbrace{R}_{0.2},\underbrace{ST}_{0.2}\} \\
    &\quad\rightarrow\quad
    \{\underbrace{P}_{0.4},\underbrace{RST}_{0.4}, \underbrace{Q}_{0.2}\} \\
    &\quad\rightarrow\quad
    \{\underbrace{QRST}_{0.6}, \underbrace{P}_{0.4}\}
    \end{align*}

2. Splitting = assign 0/1 to each split

    \begin{align*}
    \{QRST,P\}
    &\quad\rightarrow\quad
    \{\underbrace{QRST}_{0},\underbrace{P}_{1}\}\\
    &\quad\rightarrow\quad
    \{\underbrace{Q}_{00},\underbrace{RST}_{01}, \underbrace{P}_{1},\}\\
    &\quad\rightarrow\quad
    \{\underbrace{Q}_{00},\underbrace{R}_{010},\underbrace{ST}_{011}, \underbrace{P}_{1}\}\\
    &\quad\rightarrow\quad
    \{\underbrace{Q}_{00},\underbrace{R}_{010},\underbrace{S}_{0110},\underbrace{T}_{0111}, \underbrace{P}_{1}\} 
    \end{align*}
    
So one of the Huffman coding scheme is $\{P,Q,R,S,T\} \rightarrow \{1,00,010,0110,0111\}$.
