# 1. A continum of strategies

The definition of Nash equilibrium under continum of strategies is the same as in the discrete strategies case. The only thing changed is that each players can choose their strategies from a continuous space rather than discrete number of strategies. Consider an $N$ players game in general,

- The strategy $q_i$ of the $i^\text{th}$ player is a continuous number (or tuple of numbers).
- Each player's payoff function $\pi_i(q_1,q_2,...,q_N)$ is a function over the strategies $q_i$ of all players.
- The strategy profile $(q^*_1,q^*_2,...,q^*_N)$ is a Nash equilibrium if 
 
 $$\pi_i(q^*_1,...,q^*_i,...,q^*_N) \geq \pi_i(q^*_1,...,q_i,...,q^*_N)$$

 i.e. Every player are playing his best response strategy to each other. He cannot be better off if he deviates from his Nash strategy while the other players play their Nash strategies.


> **The calculus approach**
>
>In the case that **_the payoff functions $\pi_i$ have continuous second derivatives_**, a Nash equilibrium may be found by pure calculus:
>
>1. Solve the simultaneous system of equations
>
> $$\frac{\partial \pi_i}{\partial q_i} = 0 \quad\text{for all }i$$
> 
>2. Check if the solution $(q^*_1,q^*_2,...,q^*_N)$ of the above system is unique.
>
>3. Check if the solution $(q^*_1,q^*_2,...,q^*_N)$ is a maximum in all of the payoff functions, i.e. satisfy
> 
> $$\left.\frac{\partial^2 \pi_i}{\partial q_i^2}\right|_{(q^*_1,q^*_2,...,q^*_N)} <0 \quad\text{for all }i $$
> 
>
>If the solution satisfies all of the above 3 requirements, it must be a Nash equilibrium. However 
>- This approach may not find all Nash point.
>- If the payoff functions has no derivative, we need to do direct analysis.

#### Example: Guessing $\frac{2}{3}$ of the average

$3$ persons are going to choose any number $q\geq 1$. The person who chooses closest to $\frac{2}{3}$ of the average of all the numbers chosen wins $1$ dollar. If two or more people choose the same number closest to $\frac{2}{3}$ of the average, they split the $1$ dollar. The payoff function is thus

$$
\pi_i(q_1,q_2,q_3) = 
\begin{cases}
1 &\text{if }q_1 \neq q_2 \neq q_3 \text{ and }q_i\text{ is closest to }\bar{q}\\
\frac{1}{2} &\text{if }q_i = q_j \text{ for some }i\neq j\text{ and }q_i\text{ is closest to }\bar{q}\\
\frac{1}{3} &\text{if }q_1=q_2=q_3\\
0 &\text{otherwise}
\end{cases}
$$

where $\bar{q}=\frac{2}{3}\frac{q_1+q_2+q_3}{3}$ is the $\frac{2}{3}$ of the average. _The payoff function is not continuous, so we have to do direct analysis._

**Claim:** $(q_1,q_2,q_3)=(1,1,1)$ is a Nash equilibrium.

**Proof:** To show that $(1,1,1)$ is a Nash equilibrium, it suffices to show that 

$$\pi_1(1,1,1) \geq \pi_1(k,1,1) \quad\text{for any } k>1$$

i.e. If any player (let's say player 1) deviates from choosing $1$ while other players choose $1$, he gets worse payoff. By $\bar{q} = \frac{2}{3}\frac{k+1+1}{3}$, player 1 is better off only if his $k$ is closer to $\bar{q}$ then the $1$'s chosen by other players, i.e. 

$$
\begin{align*}
\left|k-\bar{q}\right| &\leq \left|\bar{q}- 1\right| \\
\left|\frac{7}{9}k-\frac{4}{9}\right| &\leq \left|\frac{2}{9}k-\frac{5}{9}\right|\\
-\frac{1}{5} \leq &\ \ k \leq 1
\end{align*}
$$

So obviously player 1 can never win if he chooses any $k>1$. Therefore 

$$\pi_1(1,1,1) =\frac{1}{3} > 0 = \pi_1(k,1,1) \quad\text{for any } k>1$$

and so $(1,1,1)$ is a Nash equilibrium.

## 2.1. Tragedy of the commons

The tragedy of the commons describes the situation when individual users, who have open access to a shared resource, act independently according to their own self-interest and contrary to the common good of all users, will cause depletion of the resource through their uncoordinated action.

#### Example: Grazing land for sheep

Suppose there are $N$ farmers who share grazing land for sheep. Each of the farmers has the option of having up to $q_i =10$ sheep. The payoff to a farmer for having a sheep is $a$ dollar, but sheep damage the common grazing land at cost $b$ dollar per sheep (shared by all the $N$ farmers).

- The strategies of each farmer can be $0\leq q_i\leq 10$
 
- The payoff of each farmer is then

 $$\pi_i(q_1,...,q_N) = aq_i-b \frac{q_1+q_2+...+q_N}{N}$$

**Claim:** The Nash equilibrium are 

$$
(q_1,q_2,...,q_N)=
\begin{cases}
(10,10,...,10) &\text{if }N\geq\frac{b}{a} \\ 
(0,0,...,0) &\text{if }N\leq\frac{b}{a}
\end{cases}
$$

**Proof:** 

1. First proof the first case. If every farmer owns a sheep, the payoff for each farmer is <br>  
$$\pi_i(10,10,...,10)=10a-10b$$  
If the farmer $i$ decides to own one fewer sheep, his payoff is <br>  
$$\pi_i(10,10,...,9,...,10)=9a-b\frac{10N-1}{N}$$  
Therefore,  $(q_1,q_2,...,q_N)=(10,10,...,10)$ is a Nash equilibrium, if <br>  
$$
\begin{align*}
\pi_i(10,10,...,10)=10a-10b &\geq 9a-b\frac{10N-1}{N} = \pi_i(10,10,...,9,...,10) \\
N &\geq \frac{b}{a}
\end{align*}
$$

2. The second case is similar. If no farmer owns a sheep, the payoff for each farmer is <br>  
$$\pi_i(0,0,...,0)=0$$  
If the farmer $i$ decides to own one sheep, his payoff is <br>  
$$\pi_i(0,0,...,1,...,)=a-b\frac{1}{N}$$  
Therefore,  $(q_1,q_2,...,q_N)=(0,0,...,0)$ is a Nash equilibrium, if <br>  
$$
\begin{align*}
\pi_i(0,0,...,0)=0 &\geq a-b\frac{1}{N} = \pi_i(0,0,...,1,...,0) \\
\frac{b}{a} &\geq N
\end{align*}
$$

Therefore when the profit per sheep is higher than the shared cost for grazing land (i.e. $a>b$, thus $N>\frac{b}{a}>1$), every farmer will maximize his number of sheep and deplete the grazing land. 

If you are the government who wants to protect the grazing land, the solution is quite simple - introduce a tax to decrease the profit per sheep, i.e. decrease $a$ to $a'$ so that $N<\frac{b}{a'}$. Then no farmer would want to keep a sheep.

## 2.2. Cournot duopoly

Cournot developed one of the earliest economic models of the competition between two firms. Suppose that there are two firms $i=1,2$ producing the same product with quantity $q_i\geq 0$, so the total quantity produced is $q = q_1+q_2$Â· Each firm seeks to maximize their individual profit given their competitors' decisions.

- Suppose the price $P$ of the product only depends on difference between the market's total supply $q$ and total demand $D$ which is a constant. For simplicity, we may consider a simple linear function: <br>  
$$P(q) = \max(k(D-q), 0)$$
 
- The payoff of the two firms are <br>  
$$
\begin{cases}
u_1(q_1,q_2) = P(q_1+q_2)q_1 - c_1q_1 \\
u_2(q_1,q_2) = P(q_1+q_2)q_2 - c_2q_2
\end{cases}
$$  
where $c_1,c_2$ are the marginal cost for producing one unit of product by the two firms. Usually, they are functions to the quantities produced by individual firms (e.g. mass production = cost less). For simplicity, we can assume them as constants as well.

_This time we have two continuous payoff functions (when $q<D$). So the differentiation approach is possible._ 

1. First solve the system of first derivatives: <br>  
$$
\begin{cases}
\displaystyle\frac{\partial u_1}{\partial q_1} = 0 \quad\Rightarrow\quad kD - 2kq_1-kq_2-c_1 = 0 \\[0.5em]
\displaystyle \frac{\partial u_2}{\partial q_2} = 0 \quad\Rightarrow\quad kD - 2kq_2-kq_1-c_2 = 0 
\end{cases}
$$  
On solving, we get <br>  
$$
q_1^* = \frac{kD+c_2-2c_1}{3k} \quad \text{and}\quad q_2^* = \frac{kD+c_1-2c_2}{3k}
$$  
Because the system is linear, the solution must be unique.

2. Check the second derivatives conditions: <br>  
$$
\begin{cases}
\displaystyle\frac{\partial^2 u_1}{\partial q_1^2} = -2k <0 \\
\displaystyle\frac{\partial^2 u_2}{\partial q_2^2} = -2k <0 \\
\end{cases}
$$  
So $(q_1^*, q_2^*)$ is a Nash equilibrium.

At the Nash point,

- Total quantities produced: <br>  
$$q^* = q_1^* + q_2^* = \frac{2kD -c_1-c_2}{3k}$$ 

- Product's price: <br>  
$$P(q_1^* + q_2^*) = \frac{kD+c_1+c_2}{3k}$$

- Profit to individual firms: <br>  
$$u_1^*(q_1^*,q_2^*) = \frac{(kD + c_2-2c_1)^2}{9k^2} \quad\text{and}\quad u_2^*(q_1^*,q_2^*) = \frac{(kD + c_1-2c_2)^2}{9k^2}$$


**Note:** In the case $kD < 2c_1 -c_2$, firm 1 has a too high cost per production, and so it is not competitive (And vice versa for firm 2). Then firm 1 should take $q_1^*=0$ and so firm 2's best response would be $q_2^* = \frac{kD-c_2}{2k}$. (Prove it yourself!)



## 2.3. The Bertrand model

The Bertrand model was formulated by Bertrand in a review of Cournot's model. Cournot argued that when firms choose quantities to produce, the equilibrium outcome involves firms pricing above marginal cost. While in Bertrand's review, he argued that if firms chose prices rather than quantities, then the competitive outcome would occur when the price equals to marginal cost.

- Suppose the market demand $D$ is a function of how the firms charge their product (at $p_1$ or $p_2$). We may assume it as a linear function <br>  
$$ D(p) = \max(k(\Gamma - p), 0)$$  
where the constant $\Gamma$ can be interpreted as a maximum price, at which nobody are willing to buy the product if the price exceeds certain level.

- About how the market is splited: 
    - If both firms charge the same price, they split the market evenly.
    - If one firm charges a lower price, it takes over the whole market. (Assume both firms have such ability.)  

  <br>The payoffs for the firms are therefore <br>  
$$u_i(p_1,p_2) = 
\begin{cases}
k(p_i-c_i)(\Gamma - p_i)  &\text{for the firm charging lower price} \\[0.5em]
\frac{k(p_i-c_i)(\Gamma - p_i)}{2} &\text{if both firms charge equal price} \\[0.5em]
0 &\text{for the firm charging higher price}
\end{cases}
$$  
where $c_1,c_2$ are the production cost for the firms.

**Claim:**

- When $c_1=c_2=c$, Nash point exists at $p_1=p_2=c$.
- When $c_1\neq c_2$, the firm with lower production cost will take over the market.

**Proof:** 

1. Suppose starting with $p_1>p_2>\max(c_1,c_2)$, firm 2 takes over the market and firm 1 gets $0$ payoff. However, firm 1 can maximize its payoff by decreasing $p_1$ to slightly below $p_2$, and so takes over the market. If this happens, firm 2 will do the same thing - decreasing $p_2$ to slightly below $p_1$ so that it takes over the market. The iteration continues until one of the firm cannot decrease its price anymore, i.e. when $p_1=c_1$ or $p_2=c_2$. 

2. If $p_1=p_2=c$, the payoff of both firms are $u_i(p_1,p_2) = \frac{k(p_i-c)(\Gamma - p_i)}{2}=0$. Both firms cannot increase his payoff if they deviate their price, since  
    <ul>
    <li>Increasing price = lose market. Payoff remains $0$.</li>
    <li>Decreasing price = get negative payoff.</li>
    </ul><br>
 Therefore $p_1=p_2=c$ is a Nash equilibrium.


3. When the production costs are different, say $c_1> c_2$, firm 2 only needs to set $p_2$ slightly smaller than $c_1$, then it can take over the whole market.



## 2.4. The Stackelberg model

The Stackelberg model consists of two firms competing in quantity of production in turns:

1. The leader firm always announces its strategy first.
2. The follower firms decides its best response strategy sequentially by observing leader's strategy. 

The essence of Stackelberg model is that the leader firm can predict the best response strategy played by the follower firms, and so plan the strategy to be announced so that its payoff is maximized. For simplicity, let's use the payoff function in Cournot model.

$$
\begin{cases}
u_1(q_1,q_2) = P(q_1+q_2)q_1 - c_1q_1 \\
u_2(q_1,q_2) = P(q_1+q_2)q_2 - c_2q_2
\end{cases}
$$

1. For firm 2 to maximize its payoff, it has to decide is quantity to produce $q_2$ as a function of $q_1$. The maximum occurs when <br>  
$$
\frac{\partial u_2}{\partial q_2} = 0 \quad\Rightarrow\quad q_2^*(q_1) = \frac{kD - kq_1-c_2}{2k}
$$

2. Substituting it into firm 1's payoff function, it now depends entirely on $q_1$. <br>  
$$
u_1(q_1,q_2^*(q_1)) = P(q_1+q_2^*(q_1))q_1 - c_1q_1 = q_1\left(\frac{kD-kq_1+c_2-2c_1}{2k}\right)
$$  
Then the maximum occurs when <br>  
$$
\frac{\partial u_1}{\partial q_1} = 0 \quad\Rightarrow\quad q_1^* = \frac{kD-2c_1+c_2}{2k}
$$  
The corresponding production quantity by firm 2 is then <br>  
$$
q_2^* = \frac{kD+2c_1-3c_2}{4k}
$$

At $(q_1^*,q_2^*)$, 

- Total quantity produced:  
$$q^* = q_1^*+q_2^* = \frac{3kD-2c_1-c_2}{4k}$$

- Product's price:  
$$P(q^*) = \frac{kD+2c_1+c_2}{4k}$$ 

- Payoffs for the firms are respectively: <br>  
$$
u_1(q_1^*, q_2^*) = \frac{(kD+c_2-2c_1)^2}{8k^2} \quad\text{and}\quad u_2(q_1^*,q_2^*) = \frac{(kD+2c_1-3c_2)^2}{16k^2}
$$

**Note 1:** When $c_1=c_2$, we may compare with the Cournot model,

||Cournot||Stackelberg|
|:---:|:---:|:---:|:---:|
|Total qauntity produced|$\frac{2(kD-c)}{3k}$|$<$|$\frac{3(kD-c)}{4k}$|
|Product's price|$\frac{kD+3c}{4k}$|$<$|$\frac{kD+2c}{3k}$|
|Firm 1's profit|$\frac{(kD-c)^2}{9k^2}$|$<$|$\frac{(kD-c)^2}{8k^2}$|
|Firm 2's profit|$\frac{(kD-c)^2}{9k^2}$|$>$|$\frac{(kD-c)^2}{16k^2}$|

which means in Stackelberg's situation, more products are produced at a lower price. So the social welfare is enhanced. 

**Note 2:** Suppose firm 1 has a different objective - instead of maximizing its own profit, it wants to prevent firm 2 entering the market. In order to do so, firm 1 needs to make firm 2 unprofitable. By looking at firm 2's payoff function as a function of $q_1$:

$$
u_2(q_1,q_2^*(q_1)) = P(q_1+q_2^*(q_1))q_2 - c_2q_2 = \frac{(kD-kq_1-c_2)^2}{4k^2}
$$

equals to or smaller than $0$ when $q_1\geq D-\frac{c_2}{k}$. This is the quantity firm 1 needs to produce in order to maintain its monopoly.
