This note is a summary of some famous types of 2-by2 symmetric matrix games. In general, the payoff bi-matrix is expressed as 

$$
\left(\begin{array}{c|cc}
&B_1 &B_2 \\
\hline
A_1 &(r,r) &(s,t) \\
A_2 &(t,s) &(u,u)
\end{array}\right)
$$

The game is classified based on the size of $r,s,t,u$.

|Classification|Relation between $r,s,t,u$|
|:---:|:---:|
|Prisoner's dilemma|$s<u<r<t$|
|Stag hung|$s<u<t<r$|
|Pure coordination|Trivial: $s<t<r<u$ <br> Non-trivial: $s<t<u<r$|
|Chicken game|$u<s<t<r$|
|Battle of sexes|$r<u<s<t$|
|Leader|$u<r<s<t$|


[Here](http://dragice.fr/gameicons/) is a good index page for visualizing different types of 2-by-2 matrix games, including asymmetric games.

# 1. Coordination games

A coordination game attains its pure Nash equilibria when both players use the same strategy.


## 1.1 Prisoner's dilemma ($s<u<r<t$)

$A$ and $B$ are arrested for a crime. The police lack evidence to charge them, so they interrogate $A$ and $B$ independently if the other person has committed the crime. They can remain silent or betray the other.

- If they both remain silent, they will be lightly imprisoned for $1$ year for the lack of evidence.
- If one remains silent and one betrays, the betrayer is immediately released, whereas the other person will be heavily imprisoned for $5$ years.
- If they both betray, they will be imprisoned for $3$ years, shorter than five years as a reward.

The game attains its only pure equilibrium when they both betray, but the situation is certainly better if they both remain silent.

$$
(\mathbf{A}, \mathbf{B}^\intercal) = 
\left(\begin{array}{c|cc} 
& \text{silent} & \text{betray} \\
\hline
\text{silent} & (-1,-1) & (-5,0) \\
\text{betray} & (0,-5) & (-3,-3)
\end{array}\right)
$$
 

## 1.2. Stag hunt ($s<u<t<r$)

$A$ and $B$ are hunters. They know that a valuable stag resides in a forest. It is so strong that it cannot be killed without two people's efforts. They can join each other to hunt the stag or hunt hares, which are less valuable, alone instead.

- If they both hunt the stag, they kill it successfully and earn $5$ units each.
- If one hunts the stag and one hunts hares, the hare hunter earns $2$ units safely, whereas the stag hunter fails to kill the stag and earns nothing.
- If they both hunt hares, they can just earn $1$ unit each, less than a single hare hunter's payoff because they have to share the hares.

The game attains its payoff-dominant equilibrium when they both hunt stags, but should they trust their partners? What if their partner is a coward and favours hunting hares safely?

$$
(\mathbf{A}, \mathbf{B}^\intercal) = 
\left(\begin{array}{c|cc} 
& \text{stag} & \text{hare} \\
\hline
\text{stag} & (5,5) & (0,2) \\
\text{hare} & (2,0) & (1,1)
\end{array}\right)
$$
 


## 1.3. Pure coordination ($s<t<r<u$ or $s<t<u<r$)

$A$ and $B$ represent two companies, competing over the use of a certain industrial standard. They can adopt standard 1 or standard 2, which is intrinsically better.

- If they adopt the same standard, customers feel more comfortable, so it is the most beneficial to both companies.
    - If they both choose standard 1, they earn $3$ units each.
    - If they both choose standard 2, they earn $4$ units each.


- If they adopt different standards, customers feel troublesome to buy adapters, so both companies earn less. But who earns more? 

This can be further divided into a trivial case and a non-trivial case.

- **Trivial case:** Because standard 1 is inferior, standard 1's user earns $1$ unit, whereas standard 2's user earns $2$ units.

 $$
 (\mathbf{A}, \mathbf{B}^\intercal) = 
\left(\begin{array}{c|cc} 
& \text{std. 1} & \text{std. 2} \\
\hline
\text{std. 1} & (3,3) & (1,2) \\
\text{std. 2} & (2,1) & (4,4)
\end{array}\right)
$$

- **Non-trivial case:** Despite its intrinsic inferiority, standard 1 is an older standard to which many customers have accustomed. Because the customers refuse to change their habits, standard 1's user earns $2$ units, whereas standard 2's user earns $1$ units.

 $$
 (\mathbf{A}, \mathbf{B}^\intercal) =
\left(\begin{array}{c|cc}
& \text{std. 1} & \text{std. 2} \\
\hline
\text{std. 1} & (3,3) & (2,1) \\
\text{std. 2} & (1,2) & (4,4)
\end{array}\right)
$$

In both cases, the game attains its payoff-dominant pure equilibrium when both $A$ and $B$ adopt standard 2. However in the non-trivial case, they have the incentive to switch to standard 1 to earn more than the other - this can be rational when the companies only care about their relative wealth. Therefore, the equilibrium is unstable in the non-trivial case.

**Note:** In generalization. The trivial case of pure coordination satisfies $s<t<r<u$, while the non-trivial case satisfies $s<t<u<r$. (The non-trivial case's matrix is reordered to satisfy $s<t$.) Hence, the triviality depends on the relative payoffs when players deviate from the payoff-dominant equilibrium.

$$
{\text{trivial}\atop\text{pure coord.}} \to
\left(\begin{array}{c|cc}
& \text{std. 1} & \text{std. 2} \\
\hline
\text{std. 1} & (3,3) & (1,2) \\
\text{std. 2} & (2,1) & (4,4)
\end{array}\right)
%
\qquad\qquad
%
{\text{non-trivial}\atop\text{pure coord.}} \to
\left(\begin{array}{c|cc} 
& \text{std. 2} & \text{std. 1} \\
\hline
\text{std. 2} & (4,4) & (1,2) \\
\text{std. 1} & (2,1) & (3,3)
\end{array}\right)
$$

# 2. Non-coordination games

A non-coordination game attains its pure Nash equilibria when $A$ and $B$ use different strategies.

 

## 2.1. Chicken game ($u<s<t<r$)

$A$ and $B$ drive on a single-lane bridge along opposite directions. There is not enough space for them to pass together, so one car must go back and leave the bridge first. They can keep going forwards or go backwards.

- If they both go forwards, their cars crash, so they both lose $5$ units for repairing the car.
- If one goes forwards and one goes backwards, the forward driver takes advantage of and thus wins $1$ unit from the backward driver.
- If they both go backwards, no one earns as the problem remains unsolved.

The game attains its pure equilibria when one goes forward and one goes backward. However, who is willing to sacrifice himself and suffer the loss?

$$
(\mathbf{A}, \mathbf{B}^\intercal) = 
\left(\begin{array}{c|cc}
& \text{back.} & \text{for.} \\
\hline
\text{back.} & (0,0) & (-1,1) \\
\text{for.} & (1,-1) & (-5,-5)
\end{array}\right)
$$
 


## 2.2. Battle of the sexes ($r<u<s<t$)

$A$ and $B$ are dating, but they cannot decide what to do. Alice wants to watch an opera at a theatre, whereas Bob wants to watch a football match at a stadium. They decide to resolve the situation with destiny: they go to one place separately and see if they can meet each other. They can choose to be selfish or altruistic (i.e. selfless).

- If one is selfish and one is altruistic, they meet each other and date happily, and the selfish one enjoys his activity as well. Hence, the selfish one earns $5$ units, whereas the selfless one earns $3$ units.
- If they are both selfish, they cannot meet each other, but they can still enjoy their activity alone, so they both earn $1$ unit.
- If they are both altruistic, not only they fail to meet each other, but they also cannot do what they want to do, so no one earns.

Similar to chicken, this game attains its pure equilibria when one is selfish and one is altruistic. But in contrast, $A$ and $B$ get hurt the most when they are both altruistic. (They would get hurt the most when they are both selfish if the game belonged to chicken game.)

$$
(\mathbf{A}, \mathbf{B}^\intercal) = 
\left(\begin{array}{c|cc}
& \text{altr.} & \text{self.} \\
\hline
\text{altr.} & (0,0) & (3,5) \\
\text{self.} & (5,3) & (2,2)
\end{array}\right)
$$


## 2.3. Leader ($u<r<s<t$)

$A$ and $B$ are teammates and both want to be the team leader. However, there can be one leader only, so one of them must give up. They decide if they should continue aiming at the leadership or give up for the other person.

- If one continues and one gives up, the former becomes the leader and earns $5$ units, whereas the latter becomes the deputy leader and earns $3$ units.
- If they both continue, they will have a quarrel and hurts the team, so they lose $2$ units each.
- If they both give up, the team still does not have a leader, so no one earns.

The game, again similar to chicken gane, attains its pure equilibria when one gives up and one continues. This is however different because the equilibria payoff-dominate the situation when they both give up.

$$
(\mathbf{A}, \mathbf{B}^\intercal) = 
\left(\begin{array}{c|cc}
& \text{give} & \text{cont.} \\
\hline
\text{give} & (0,0) & (3,5) \\
\text{cont.} & (5,3) & (-2,-2)
\end{array}\right)
$$
